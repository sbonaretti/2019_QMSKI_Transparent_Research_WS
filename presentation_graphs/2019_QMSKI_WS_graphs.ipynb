```python
%%R
# font size for all graphs
title_font_size        <- 10
label_font_size        <- 8
legend_font_size       <- 5
```

# Load data
Data collected from https://scholar.google.com and https://app.dimensions.ai/discover/publication using keyword and modality described in the table

```python
data_name = "./2019_QMSKI_WS_graphs.csv"
data = pd.read_csv(data_name)
display(data)
```

```python
# drop last four rows
data = data.drop (data.shape[0]-1, axis = 0) # Inspired by
data = data.drop (data.shape[0]-1, axis = 0) # Search modality
data = data.drop (data.shape[0]-1, axis = 0) # Keyword
data = data.drop (data.shape[0]-1, axis = 0) # Website
# make sure values are integers
data["year"]     = pd.to_numeric(data["year"])
data["python"]   = pd.to_numeric(data["python"])
data["matlab"]   = pd.to_numeric(data["matlab"])
data["figshare"] = pd.to_numeric(data["figshare"])
data["zenodo"]   = pd.to_numeric(data["zenodo"])
display(data)
```

# Use of python from scholar.google.com

```python
%%R -i data -h 2 -w 3 --units in -r 300
library(ggplot2)

plot <- ggplot(data, aes(x=year, y=python))+
        geom_point(color = '#306998') + 
        # graph ratio and limits
        coord_cartesian(ylim=c(0,80000)) +
        # background
        theme_bw() +
        # title
        ggtitle("python") +
        theme(plot.title = element_text(size = title_font_size, hjust = 0.5)) + # hjust centers title
        # axis labels 
        xlab("Year") +
        ylab("Search counts") +
        theme(text = element_text(size = label_font_size),
             axis.text.x = element_text(angle=45, hjust = 1)) +
        # axis ticks
        scale_x_continuous(breaks = seq(2003, 2019, by=1)) +
        scale_y_continuous(breaks = seq(0, 80000, by=10000)) +
        # legend off
        scale_color_discrete(guide=FALSE)
        
# display plot
plot
```

# Use of Zenodo and Figshare from app.dimensions.ai

```python
# remove cells with NA
data = data[~data.figshare.isnull()]
display(data)
```

```python
%%R -i data -h 2 -w 3 --units in -r 300
library(ggplot2)

plot <- ggplot(data, aes(x=year, y=zenodo+figshare, color = '#1b9e77'))+ # colors have to be in aes to get a legend (colorblind safe)
        geom_point() + 
        geom_point(aes(x=year, y=zenodo, color='#7570b3')) +
        geom_point(aes(x=year, y=figshare, color='#d95f02')) +  
        # colors
        scale_colour_manual(values = c("#1b9e77", "#7570b3", "#d95f02"), #green, violet, orange 
                            labels = c(" Zenodo+Figshare", " Zenodo", " Figshare")) + # restating the colors for the legend 
        # graph ratio and limits
        coord_cartesian(ylim=c(0,13000)) +
        # background
        theme_bw() +
        # title
        ggtitle("Zenodo and Figshare") +
        theme(plot.title = element_text(size = title_font_size, hjust = 0.5)) + # hjust centers title
        # axis labels 
        xlab("Year") +
        ylab("Search counts") +
        theme(text = element_text(size = label_font_size),
             axis.text.x = element_text(angle=45, hjust = 1)) +
        # axis ticks
        scale_x_continuous(breaks = seq(2009, 2019, by=1)) +
        scale_y_continuous(breaks = seq(0, 13000, by=2000)) +
        # legend 
        theme(legend.position = c(0.2,0.8)) + 
        theme(legend.text=element_text(size=legend_font_size)) + # text size
        theme(legend.key.size = unit(0.05,"in")) +  # legend size
        theme(legend.title=element_blank())  # remove legend title

# display plot
plot
```
